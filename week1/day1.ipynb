```python
from dotenv import load_dotenv

load_dotenv(override=True)
api_key = os.getenv("GOOGLE_API_KEY")

if not api_key: 
    raise ValueError("GOOGLE_API_KEY environment variable not set")
elif api_key.strip() != api_key:
    raise ValueError("GOOGLE_API_KEY environment variable contains leading or trailing whitespace") 
else:
    print("GOOGLE_API_KEY environment variable is set and valid")
```

```python
from google import generativeai as genai
genai.configure(api_key=api_key)

import pprint
for model in genai.list_models():
    pprint.pprint(model)
model = genai.GenerativeModel('models/gemini-2.5-flash')
result = model.generate_content('Tell me a story about a magic backpack')
result.text
```
