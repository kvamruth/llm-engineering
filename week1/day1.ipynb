```python
import os
from dotenv import load_dotenv
from scraper import fetch_Website_contents
from IPython.display import Markdown, display
import google.generativeai as genai
```

```python
from dotenv import load_dotenv

load_dotenv()

api_key = os.environ.get("AI_API_KEY")
# api_key = os.getenv("GOOGLE_API_KEY")

if api_key:
    api_key = api_key.strip()
    if not api_key:
        print("GOOGLE_API_KEY is empty after stripping")
    else:
        print("GOOGLE_API_KEY environment variable is set and valid")
else:
    print("GOOGLE_API_KEY environment variable not set")
```

```python
from google import generativeai as genai
genai.configure(api_key=api_key)

import pprint
for model in genai.list_models():
    pprint.pprint(model)
model = genai.GenerativeModel('models/gemini-3.1-flash-lite-preview')
result = model.generate_content('Tell me a story about a magic backpack')
result.text
```

```python
# url = 'https://www.crummy.com/software/BeautifulSoup/'
url = "https://anthropic.com"
sc = fetch_Website_contents(url)
print(type(sc))
print(sc)
```

```python
system_prompt = "You are an expert in web scraping and data extraction. Your task is to analyze the provided website content and extract key information, such as the main purpose of the website, its target audience, and any unique features it offers. Please provide a concise and humorous summary of your findings in pig latin.If it contains any text related to navigation, please ignore it and focus only on main content of the web.Respond in markdown format."
```

```python
user_prompt = """
Here are the contents of the website.
Provide a concise summary of the website in markdown,
If it includes any news or announcements , please include them in the summary.   
"""
```

```python
messages = [{
    "role": "system",
    "parts": [system_prompt]
}, {
    "role": "user",
    "parts": [user_prompt]
}]
response = model.generate_content(messages)
```

```python
def msgs_for(webcontent):
    return [{
        "role":"system",
        "content":system_prompt
    },{
        "role":"user",
        "content":user_prompt+webcontent
    }]
```

```python
from openai import OpenAI
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key
)
msgs = msgs_for(sc)
# print(msgs)
response = client.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=msgs
)
display(Markdown(response.choices[0].message.content))
```

```python
import requests

msgs = [{
    "role":"user",
    "content":"Tell me few fun facts about space and universe"
}]

headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type":"application/json"
}

payload = {
    "model":"openai/gpt-oss-20b:free",
    "messages":msgs
}

res = requests.post(
    "https://openrouter.ai/api/v1/chat/completions",
    headers=headers,
    json=payload
)
# res.json()
print(res.json()["choices"][0]["message"]["content"])
```
