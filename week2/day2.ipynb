```python
from openai import OpenAI
import os
api_key = os.environ.get("AI_API_KEY")
if api_key:
    print("API KEY IS VALID")
else:
    raise ValueError("API key is not valid and set")

openai = OpenAI(
    api_key=api_key,
    base_url="https://openrouter.ai/api/v1"
)
```

```python
sys_prompt = """
You are a helpful assistant, Who answers any query by user in a short and informative manner
"""
def msg_gpt(prompt):
    response = openai.chat.completions.create(
        model="openai/gpt-oss-20b:free",
        messages=[
            {"role":"system","content":sys_prompt},
            {"role":"user","content":prompt},
        ]

    )
    return response.choices[0].message.content
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.environ.get("AI_API_KEY")
# api_key = os.getenv("GOOGLE_API_KEY")

if api_key:
    api_key = api_key.strip()
    if not api_key:
        print("GOOGLE_API_KEY is empty after stripping")
    else:
        print("GOOGLE_API_KEY environment variable is set and valid")
else:
    print("GOOGLE_API_KEY environment variable not set")
```

```python
system_prompt = "You are an expert in web scraping and data extraction. Your task is to analyze the provided website content and extract key information, such as the main purpose of the website, its target audience, and any unique features it offers. Please provide a concise summary of your findings. If it contains any text related to navigation, please ignore it and focus only on main content of the web.Respond in markdown format."
```

```python
from openai import OpenAI
def stream_ai(sys_prompt,user_prompt):
    client = OpenAI(
        base_url="https://openrouter.ai/api/v1",
        api_key=api_key
    )
    msgs = [
        {"role":"system","content":sys_prompt},
        {"role":"user","content":user_prompt}
    ]
    response = client.chat.completions.create(
        model="openai/gpt-oss-20b:free",
        messages=msgs,
        stream=True
    )
    result = ""
    for chunk in response:
        delta = chunk.choices[0].delta.content or ''
        result += delta
        yield result
```

```python
from scraper_util import fetch_Website_contents

def stream_brochure(comp_name,comp_url):
    prompt = f"Please generate a company brochure for {comp_name}, Here is their landing page:\n"
    prompt += fetch_Website_contents(comp_url)
    yield from stream_ai(sys_prompt=system_prompt,user_prompt=prompt)
```

```python
import gradio as gr
name_input = gr.Textbox(label="Company Name")
url_input = gr.Textbox(label="Landing page url including http:// or https://")
msg_output = gr.Markdown(label="Response")

view = gr.Interface(
    fn=stream_brochure,
    title="AI Brochure Generator",
    inputs=[name_input,url_input],
    outputs=[msg_output],
    examples=[["Hugging Face","https://huggingface.co"]],
    flagging_mode="never",
    theme=gr.Theme.from_hub("Bruhn/CrimsonNight")
)
view.launch()
```
